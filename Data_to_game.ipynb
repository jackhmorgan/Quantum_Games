```python
import numpy as np

def random_strategy_market(n_turn):
    stonks = [1]
    changes_in_price = [-0.2, 0.05, 0.3]
    probs = [0.25, 0.6, 0.15]
    for n in range(n_turn):
        change = np.random.choice(changes_in_price, p=probs)
        stonks.append(stonks[-1]+change)
    return stonks
```
